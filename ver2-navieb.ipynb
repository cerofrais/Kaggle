```python
df_train= pd.read_csv('./Accident_train.csv')
df_test=pd.read_csv('./Accident_test.csv')
```

```python
# special conditions at site has 75% as 1 so replacing with mode.
df_train['Special_Conditions_at_Site'].fillna(1,inplace=True)
df_test['Special_Conditions_at_Site'].fillna(1,inplace=True)

# junction replacing with 6 (cross roads)
df_train['Junction_Detail'].fillna(6,inplace=True)
df_test['Junction_Detail'].fillna(6,inplace=True)

#Junction controlrandomly fill 1 or 7
df_train['Junction_Control'].fillna(np.random.choice([1,7]),inplace=True)
df_test['Junction_Control'].fillna(np.random.choice([1,7]),inplace=True)

#pedcrossing by 1
df_train['Ped_Crossing_HC'].fillna(1,inplace=True)
df_train['Ped_Crossing_PC'].fillna(1,inplace=True)
df_test['Ped_Crossing_HC'].fillna(1,inplace=True)
df_test['Ped_Crossing_PC'].fillna(1,inplace=True)

#road surface 1,2
df_train['Road_Surface_Conditions'].fillna(np.random.choice([1,7]),inplace=True)
df_test['Road_Surface_Conditions'].fillna(np.random.choice([1,7]),inplace=True)

#hour of collsion  random between 12 to 19
df_train['Hour_of_Collision'].fillna(np.random.randint(12,19),inplace=True)
df_test['Hour_of_Collision'].fillna(np.random.randint(12,19),inplace=True)

# Policing areas 
df_train['Policing_Area'].fillna(np.random.choice(['BELC','ARBC','NEMD','LISC','LISB']),inplace=True)
df_test['Policing_Area'].fillna(np.random.choice(['BELC','ARBC','NEMD','LISC','LISB']),inplace=True)

```

```python
nan_rows = df_test[df_test.isnull().T.any().T]
nan_rows
```

```python
df_train=df_train.drop(['Collision_Ref_No'],axis=1)
df_train=df_train.drop(['Policing_Area'],axis=1)
df_test=df_test.drop(['Collision_Ref_No'],axis=1)
df_test=df_test.drop(['Policing_Area'],axis=1)
```

```python
#onehot_pa=pd.get_dummies(df_train['Policing_Area'])
#onehot_week=pd.get_dummies(df_train['Weekday_of_Collision'])
#
#df_train=pd.concat([df_train,onehot_week],axis=1)#onehot_pa,
#
#onehot_week=pd.get_dummies(df_test['Weekday_of_Collision'])
#
#df_test=pd.concat([df_test,onehot_week],axis=1)#onehot_pa,

df_train=df_train.replace({'Weekday_of_Collision' : { 'MON' : 0, 'TUE' : 0, 'WED' : 0,'THU':0,'FRI':0,'SAT':1,'SUN':1 }})
df_test=df_test.replace({'Weekday_of_Collision' : { 'MON' : 0, 'TUE' : 0, 'WED' :0,'THU':0,'FRI':0,'SAT':1,'SUN':1 }})
#enc = OneHotEncoder()
#df_train=enc.transform(df_train)
df_train
```

```python
df_train=df_train.drop(['Weekday_of_Collision'],axis=1)#'Policing_Area',
df_test=df_test.drop(['Weekday_of_Collision'],axis=1)#'Policing_Area',

df_train=df_train.drop(['Ped_Crossing_HC','Weekday_of_Collision','Day_of_Collision','Month_of_Collision'],axis=1)#'Policing_Area',
df_test=df_test.drop(['Ped_Crossing_HC','Weekday_of_Collision','Day_of_Collision','Month_of_Collision'],axis=1)#'Policing_Area',
```

```python
df_train_y=df_train['Collision_Severity']
df_train=df_train.drop(['Collision_Severity'],axis=1)
df_test=df_test.drop(['Collision_Severity'],axis=1)
#df_train=df_train.drop(['Special_Conditions_at_Site'],axis=1)
df_test=df_test.drop(['Special_Conditions_at_Site'],axis=1)
```
